# Now using GD

```python
def step_gradient(points, learning_rate, m):
    m_slope = np.zeros_like(m)
    M = len(points)
    for i in range(M):
        x = points[i, :-1]
        x = np.append(x,1)
        y = points[i, -1]
        s = np.dot(x, m)
        h = 1 / (1 + np.exp(-s))
        for j in range(len(x)):
            m_slope[j] += (-1/M) * (y - h) * x[j]
    new_m = m - learning_rate*m_slope    
    return new_m
```

```python
def gd(points, learning_rate, num_iterations):
    m = np.zeros(points.shape[1])
    #c = 0
    for i in range(num_iterations):
        m  = step_gradient(points, learning_rate, m )
        print(i, " Cost: ", cost(points, m))
    return m
```

```python
def cost(points, m):
    total_cost = 0
    M = len(points)
    for i in range(M):
        x = points[i, :-1]
        x = np.append(x,1)
        y = points[i, -1]
        s = np.dot(x, m)
        h = 1 / (1 + np.exp(-s))
        total_cost += (-1/M) * (y * np.log(h) + (1-y) * np.log(1-h))
    return total_cost
```

```python
def predict(x3, m):
    M = len(x3)
    y_p = []
    x3 = np.c_[ x3, np.ones(M) ]
    for i in range(M):
        s = np.dot(x3[i,:], m)
        h = 1 / (1 + np.exp(-s))
        if h > 0.5:
            y_p.append(1)
        else:
            y_p.append(0)
    return y_p
```

```python
def score(y_true, y_pred):
    accuracy = np.sum(y_true == y_pred) / len(y_true)
    return accuracy
```

```python
learning_rate = 0.001
num_iterations = 1000
m  = gd(data_train, learning_rate, num_iterations)
print("m: ",m)

y_test_pred = predict(x_test,m)
print("Test Score: ", score(y_test, y_test_pred))  #accuracy

y_train_pred = predict(x_train, m)
print("Train Score: ", score(y_train, y_train_pred))
```
